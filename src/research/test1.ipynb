```python
import os
from dotenv import load_dotenv
from cerebras.cloud.sdk import Cerebras

load_dotenv()

video_title = """



How OOP works

"""

client = Cerebras(
    api_key=os.environ.get("CEREBRAS_API_KEY"),
)

chat_completion = client.chat.completions.create(
    messages=[
        {
            "role": "user",
            "content": f"""Is the youtube video '{video_title}' educational. Give your reasoning and then respond with 'YES' or 'NO'.
            I am a university student trying to better myself, and I want to avoid distractions. 
            Say yes to videos which will help me directly improve myself. 
            I am currently studying software engineering, AI, Machine learning and am interested in startups or making money. Use this for context but this does not mean approve all videos in these niches. only the directly relevant ones.
            
            """,
        }
    ],
    model="llama3.1-8b", 
)

print(chat_completion.choices[0].message.content)
```
